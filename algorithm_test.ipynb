```python
tree = ba.Tree()
tree.make_tree(["A", "B", "C", "D", "E", "F", None, "G"])
print("전위 순회 결과: ", tree.preorder())
print("중위 순회 결과: ", tree.inorder())
print("후위 순회 결과: ", tree.postorder())
print("레벨 순회 결과: ", tree.levelorder())
tree.insert_key("H")
print("키 삽입후 레벨 순회 결과: ", tree.levelorder())
tree.delete_key("B")
print("키 삭제후 레벨 순회 결과: ", tree.levelorder())
print("형제가 없는 노드: ", tree.find_has_no_sibling())
```
