# 코딩도장_다음입사문제

1차원의 점들이 주어졌을 때, 그 중 가장 거리가 짧은 것의 쌍을 출력하는 함수를 작성하시오. (단 점들의 배열은 모두 정렬되어있다고 가정한다.)

예를들어 S={1, 3, 4, 8, 13, 17, 20} 이 주어졌다면, 결과값은 (3, 4)가 될 것이다.

```python
s = [1, 3, 4, 8, 13, 17, 20]
ss = [3, 4, 8, 13, 17, 20]
sorted(list(zip(s, ss)), key=lambda i:i[1]-i[0])[0]
```

# 2018_카카오_비밀지도_난이도_하

네오는 평소 프로도가 비상금을 숨겨놓는 장소를 알려줄 비밀지도를 손에 넣었다. 그런데 이 비밀지도는 숫자로 암호화되어 있어 위치를 확인하기 위해서는 암호를 해독해야 한다. 다행히 지도 암호를 해독할 방법을 적어놓은 메모도 함께 발견했다.

1. 지도는 한 변의 길이가 n인 정사각형 배열 형태로, 각 칸은 “공백”(“ “) 또는 “벽”(“#”) 두 종류로 이루어져 있다.
2. 전체 지도는 두 장의 지도를 겹쳐서 얻을 수 있다. 각각 “지도 1”과 “지도 2”라고 하자. 지도 1 또는 지도 2 중 어느 하나라도 벽인 부분은 전체 지도에서도 벽이다. 지도 1과 지도 2에서 모두 공백인 부분은 전체 지도에서도 공백이다.
3. “지도 1”과 “지도 2”는 각각 정수 배열로 암호화되어 있다.
4. 암호화된 배열은 지도의 각 가로줄에서 벽 부분을 1, 공백 부분을 0으로 부호화했을 때 얻어지는 이진수에 해당하는 값의 배열이다.




|n|arr1|arr2|output|
|:---|:---|:---|:---|
|5|[9, 20, 28, 18, 11]|[30, 1, 21, 17, 28]|["#####","# # #", "### #", "# ##", "#####"]
|6|[46, 33, 33 ,22, 31, 50]|[27 ,56, 19, 14, 14, 10]|["######", "### #", "## ##", " #### ", " #####", "### # "]|

```python
def solution(n, arr1, arr2):
    result = []
    for idx in range(n):
        number1 = arr1[idx]
        binary1 = [ '0' for _ in range(n) ]

        number2 = arr2[idx]
        binary2 = [ '0' for _ in range(n) ]

        if number1 > 0:
            in_idx1 = n-1
            while True:
                if number1<2:
                    binary1[in_idx1] = '1'
                    break
                binary1[in_idx1] = str(number1%2)
                number1 = int(number1/2)
                in_idx1 -= 1

        if number2 > 0:
            in_idx2 = n-1
            while True:
                if number2<2:
                    binary2[in_idx2] = '1'
                    break
                binary2[in_idx2] = str(number2%2)
                number2 = int(number2/2)
                in_idx2 -= 1

        result.append(''.join(list(map(lambda x: ' ' if '00' == x[0]+x[1] else '#', list(zip(binary1, binary2))))))

    return result
```
