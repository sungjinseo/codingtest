```python
tree = ba.Tree()
tree.make_tree(["A", "B", "C", "D", "E", "F", None, "G"])
print("  : ", tree.preorder())
print("  : ", tree.inorder())
print("  : ", tree.postorder())
print("  : ", tree.levelorder())
tree.insert_key("H")
print("    : ", tree.levelorder())
tree.delete_key("B")
print("    : ", tree.levelorder())
print("  : ", tree.find_has_no_sibling())
```
